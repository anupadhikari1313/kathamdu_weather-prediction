```python
import pandas as pd
```

```python
df=pd.read_csv('climate_data_nepal_district_wise_daily_part_i.csv',index_col='DATE')
df
```

```python
climate=df.loc[(df.DISTRICT=="Kathmandu")]
climate
```

```python
climate.columns
```

```python
climate=climate[['PRECTOT','T2M_MAX','T2M_MIN']]
```

```python
climate=climate.rename(columns={'T2M_MAX':'temp_max','T2M_MIN':'temp_min','PRECTOT':'precip'})
```

```python
climate.head()
```

```python
from datetime import datetime
climate.index=pd.to_datetime(climate.index)
climate
```

```python
climate.isnull().sum()
```

```python
climate.dtypes
```

```python
climate[["temp_max", "temp_min"]].plot()
```

```python
climate["precip"].plot()
```

```python
climate["target"] = climate.shift(-1)["temp_max"]
climate
```

```python
temp = climate.iloc[:-1,:].copy()
temp
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
predictors = ["precip", "temp_max", "temp_min"]
```

```python
train = temp.loc[:"2018-12-31"]
test = temp.loc["2019-01-01":]
```

```python
train
```

```python
test
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test["target"], predictions)
```

```python
results = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
results.columns = ["actual", "predictions"]
```

```python
results
```

```python
results.plot()
```

```python
reg.score(train[predictors], train["target"])
```

```python
reg.coef_
```
